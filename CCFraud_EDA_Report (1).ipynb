```python
# EDA Report on the European Credit Craud Fraud Dataset (Kaggle)
# By: Sanjay Saini


# Import  libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/creditcard.csv')

# Check first few rows
df.head()

# Basic info about the dataset
df.info()

# Descriptive stats
df.describe()
```

```python
# Check for any missing values

df.isnull().sum()
```

```python
# Class distribution

sns.countplot(x='Class', data=df)
plt.title('Class Dist')
plt.show()

# Print percentage of fraudulent transactions
fraud_percentage = (df['Class'].value_counts()[1] / len(df)) * 100
print(f"Fraudulent transactions: {fraud_percentage:.3f}%")
```

```python
# Correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Since PCA is designed to minimize
# multicollinearity, most of features aren’t strongly correlated with
# each other.
```

```python
# Plot histograms for each feature

df.drop(['Class'], axis=1).hist(figsize=(20, 20), bins=50)
plt.show()
```

```python
# Analyzing the transaction amount and its distribution for both fraudulent and
# non-fraudulent transactions.


# Compare transaction amount for fraud and non-fraud cases
plt.figure(figsize=(12, 6))

sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Transaction Amount by Class')
plt.show()
```

```python
# Since there is a "Time" column that represents time elapsed since the
# first transaction, we can analyze if fraud transactions tend
# to occur during specific times.

# ** Note: Time does not correspond to absolute timestamps  but rather the
# relative time from the start of the dataset collection.
# It is a sequence of transactions across all customers.
# Dataset is an aggregation of many people's transactions, and the Time
# column represents the overall order of transactions from the dataset's start,
# not meaningful time intervals for individuals.

# From this graph, it's clear that non-fraud transactions dominate across all
# times, which makes it difficult to draw any obvious conclusions.


# Plot distribution of time for both classes
plt.figure(figsize=(12, 6))

sns.histplot(df[df['Class'] == 0]['Time'], color='blue', label='Non-Fraud', kde=True, bins=50)
sns.histplot(df[df['Class'] == 1]['Time'], color='red', label='Fraud', kde=True, bins=50)
plt.legend()
plt.title('Transaction Time Dist by Class')
plt.show()

```

```python
# Using boxplot to check for outliers in Amount feature

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Amount'])
plt.title('Transaction Amount - Outlier Detection')
plt.show()
```

```python
# Summary statistics for fraudulent transactions
fraud = df[df['Class'] == 1]
print(fraud.describe())

# Summary statistics for non-fraudulent transactions
non_fraud = df[df['Class'] == 0]
print(non_fraud.describe())


```

```python
# Variance plots for PCA features

pca_columns = [f'V{i}' for i in range(1, 29)]
# Calculate variance of each PCA component
variances = df[pca_columns].var().sort_values()

# Plot the variances
plt.figure(figsize=(10, 6))
variances.plot(kind='bar')
plt.title('Variance of PCA Components (V1 to V28)')
plt.show()

# Check components with very low variance
print("Low variance PCA components:")
print(variances[variances < 0.01])

# Based on the variance plot of the PCA components (V1 to V28), we can observe
# that certain components, such as V28, V27, V26, V25, V24, V23, and V22, have
# relatively low variance compared to others. These components might contribute
# less to the overall predictive power of models since their variance is
# quite small.
```

```python
# List of PCA features (V1 to V28)
pca_columns = [f'V{i}' for i in range(1, 29)]

# Create a plot for each PCA comp comparing Fraud vs Non-Fraud
plt.figure(figsize=(16, 100))  # Adjust heightss for multiple plots
for i, col in enumerate(pca_columns, 1):
    plt.subplot(14, 2, i)  # Create subplots with 2 columns
    sns.histplot(df[df['Class'] == 0][col], color='blue', label='Non-Fraud', kde=True, bins=50)
    sns.histplot(df[df['Class'] == 1][col], color='red', label='Fraud', kde=True, bins=50)
    plt.title(f'{col} Distribution by Class')
    plt.legend()
    plt.tight_layout()

plt.show()


# Heavy Overlap:

# Across most features, the distributions of fraud (red) and non-fraudulent
#  (blue) transactions show significant overlap.
# This overlap suggests that these individual features alone might not be very
# strong in differentiating between fraud and non-fraud. However, the combination
#  is still likely informative.


```
